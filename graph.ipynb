```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
 
```

```python
results = pd.read_csv(r'E:\data sciences\sharon\results.csv')
results
```

```python
plt.figure(figsize=(10,10))
plt.rcParams.update({'font.size':15})
plt.plot(results['epoch'], results['train/cls_loss'],linewidth=3,color='salmon')
plt.title('A graph of class training loss against Epochs',color='tomato')
plt.xlabel('Epochs',color='peru')
plt.grid(color='lightgray', linestyle='--', linewidth=1)
plt.ylabel("Class loss",color='yellowgreen')
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 10))
plt.rcParams.update({'font.size': 15})

line_color = 'mediumslateblue'   
title_color = 'darkorange'       
label_color = 'teal'            
plt.plot(results['epoch'], results['train/box_loss'], linewidth=3, color=line_color)

plt.ylabel('Training Loss', color=label_color)
plt.xlabel('Epochs', color=label_color)
plt.title("Epochs against Bounding Box Loss", color=title_color)

plt.grid(color='lightgray', linestyle='--', linewidth=1)


plt.show()
```
